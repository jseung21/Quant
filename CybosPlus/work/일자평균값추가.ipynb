```python
data = pd.read_csv('10years/A000020_동화약품_20090101to20190129.csv', engine='python')
```

```python
data['종가'][0:5].mean()
```

```python
listData = []
for i in range(0, len(data)):
    if len(data) - i >= 5:
        listData.append(data['종가'][i:i+5].mean())
    else:
        listData.append(0)
```

```python
data['5일평균']=listData
```
